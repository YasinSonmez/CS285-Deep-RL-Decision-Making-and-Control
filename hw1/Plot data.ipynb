```python
import matplotlib.pyplot as plt
import numpy as np
```

## BC

```python
x = np.array([1e3, 2e3, 5e3, 7e3, 1e4, 5e4, 1e5])
y = np.array([309, 1876, 3345, 5049, 5064, 4128, 4097])
std = np.array([477, 2005, 1970, 493, 908, 1944, 1966])
```

```python
plt.plot(x, y, '-', color='black', label="Walker2D")
plt.xscale('log')
plt.xlabel("Training steps per iteration")
plt.ylabel(" Eval Return")
plt.legend()
plt.title('Hyperparameter Sweep')
plt.fill_between(x, y - std, y + std, color='red', alpha=0.2)
```
